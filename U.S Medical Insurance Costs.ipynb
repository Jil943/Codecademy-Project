```python
ages = []
sexes = []
bmis = []
num_children = []
smoker_statuses = []
regions = []
insurance_charges = []
```

```python
def load_list_data(lst, csv_file, column_name):
    with open(csv_file) as csv_info:
        csv_dict = csv.DictReader(csv_info)
        
        for row in csv_dict:
            lst.append(row[column_name])
        return lst
```

```python
load_list_data(ages, 'insurance.csv', 'age')
load_list_data(sexes, 'insurance.csv', 'sex')
load_list_data(bmis, 'insurance.csv', 'bmi')
load_list_data(num_children, 'insurance.csv', 'children')
load_list_data(smoker_statuses, 'insurance.csv', 'smoker')
load_list_data(regions, 'insurance.csv', 'region')
load_list_data(insurance_charges, 'insurance.csv', 'charges')
```

```python
class PatientsInfo:
    
    def __init__(self, patients_ages, patients_sexes, patients_bmis, patients_num_children, 
                 patients_smoker_statuses, patients_regions, patients_charges):
        self.patients_ages = patients_ages
        self.patients_sexes = patients_sexes
        self.patients_bmis = patients_bmis
        self.patients_num_children = patients_num_children
        self.patients_smoker_statuses = patients_smoker_statuses
        self.patients_regions = patients_regions
        self.patients_charges = patients_charges
    
    def analyze_ages(self):
        
        total_age = 0
        
        for age in self.patients_ages:
            
            total_age += int(age)
        
        return ("Average Patient Age: " + str(round(total_age/len(self.patients_ages), 2)) + " years")
    
    def analyze_sexes(self):
        
        males = 0
        females = 0
        
        for sex in self.patients_sexes:
            
            if sex == "male":
                males += 1
            
            elif sex == "female":
                females += 1
            
        print("Count for Male: " + str(males))
        print("Count for Female: " + str(females))
    
    def unique_regions(self):
        
        unique_regions = []
        
        for region in self.patients_regions:
            
            if region not in unique_regions:
                
                unique_regions.append(region)
        
        return unique_regions
    
    def count_per_region(self):
        
        southwest_count = 0
        southeast_count = 0
        northwest_count = 0
        northeast_count = 0
        
        for region in self.patients_regions:
            
            if region == 'southwest':
                
                southwest_count += 1
            
            elif region == 'southeast':
                
                southeast_count += 1
                
            elif region == 'northwest':
                
                northwest_count += 1
                
            elif region == 'northeast':
                
                northeast_count += 1
                
        print('Count for SouthWest Region: ' + str(southwest_count))
        print('Count for SouthEast Region: ' + str(southeast_count))
        print('Count for NorthWest Region: ' + str(northwest_count))
        print('Count for NorthEast Region: ' + str(northeast_count))
        
    def smoker_count(self):
        
        smoker = 0
        non_smoker = 0
        
        for status in self.patients_smoker_statuses:
            
            if status == 'yes':
                
                smoker += 1
                
            else:
                
                non_smoker += 1
                
        return "There are " + str(smoker) + " smokers and " + str(non_smoker) + " non-smokers"
    
    def analyze_charges(self):
        
        total_charges = 0
        
        for charge in self.patients_charges:
            
            total_charges += float(charge)
        
        return ("Average Yearly Medical Insurance Charges: " + str(round(total_charges/len(self.patients_charges), 2)) + " dollars")
    
    def create_dictionary(self):
        
        self.patients_dictionary = {}
        self.patients_dictionary["age"] = [int(age) for age in self.patients_ages]
        self.patients_dictionary["sex"] = self.patients_sexes
        self.patients_dictionary["bmi"] = [float(bmi) for bmi in self.patients_bmis]
        self.patients_dictionary["children"] = [int(children) for children in self.patients_num_children]
        self.patients_dictionary["smoker"] = self.patients_smoker_statuses
        self.patients_dictionary["regions"] = self.patients_regions
        self.patients_dictionary["charges"] = [float(charges) for charges in self.patients_charges]
        
        return self.patients_dictionary
        
```

```python
patient_info = PatientsInfo(ages, sexes, bmis, num_children, smoker_statuses, regions, insurance_charges)
```

```python
patient_info.analyze_ages()
```

```python
charges_by_smoker_status = df.groupby('smoker')['charges'].median()
```

```python
charges_by_smoker_status.plot.bar(title='Median Charges by Smoker Status',xlabel='Smoker Status', ylabel='Median Charges', color=['red', 'green'], rot=0)
```

As we can see, a smoker will pay around 4 times more than a non-smoker regarding insurance charges. 

```python
df['age_range'] = pd.cut(df['age'], bins=[17, 26, 38, 51, 64], labels=['18-26', '27-38', '39-51', '52+'])
```

```python
charges_by_age = df.groupby('age_range')['charges'].median()
```

```python
charges_by_age.plot.bar(title='Median Charges by Age Group', xlabel='Age Group', ylabel='Median Charges', rot=0, colormap='viridis')
```

It looks like there is a correlation between age and insurance charges as the median charges kept rising after each age group.

```python
charges_by_gender = df.groupby(['age_range', 'sex'])['charges'].median()
```

```python
charges_by_gender.plot.bar(title='Median Charges by Age Range and Gender', xlabel='Age Range and Gender', ylabel='Median Charges', color=['pink', 'navy']*2)
```

It also looks like females tend to pay a bit more insurance charges than males

```python
df['bmi_health'] = pd.cut(df['bmi'], bins=[0, 18.5, 25.0, 30.0, 60], labels=['Underweight', 'Healthy', 'Overweight', 'Obese'])
```

```python
charges_by_bmi = df.groupby('bmi_health')['charges'].median()
```

```python
charges_by_bmi.plot.bar(title='Median Charges by BMI group', xlabel='BMI Group', ylabel='Median Charges', color=['navy', 'green', 'orange', 'red'], rot=0)
```

We can observe that a person with a higher BMI will pay more insurance charges
